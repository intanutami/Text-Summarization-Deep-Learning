```python
import pandas as pd
import re
import os
```

```python
# Mount Google Drive untuk mengakses file
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Path ke file dataset di Google Drive
file_path = "/content/drive/MyDrive/!DEEP LEARNING/DATASET/dataset.csv"
```

```python
# Muat dataset
data = pd.read_csv(file_path)
data.head()
```

```python
# Tampilkan jumlah total data
total_data = len(data)
total_data
```

```python
# Hapus kolom 'title', 'released', dan 'url' dari DataFrame
data = data.drop(columns=['title', 'released', 'url'], errors='ignore')
data.head()
```

```python
# Mengecek baris duplikat
num_duplicates = data.duplicated().sum()
print(f"Jumlah baris duplikat: {num_duplicates}")

data = data.drop_duplicates()

# Mengecek hasil
print(f"Jumlah baris setelah menghapus duplikat: {len(data)}")
```

```python
# Hapus karakter non-ASCII dari kolom 'content' dan 'summary'
data['content'] = data['content'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', ' ', str(x)))
data['summary'] = data['summary'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', ' ', str(x)))
print("\n Menghapus karakter non-ASCII")
data.head()
```

```python
# Hapus spasi ganda
data['content'] = data['content'].apply(lambda x: re.sub(r'\s+', ' ', x).strip())
data['summary'] = data['summary'].apply(lambda x: re.sub(r'\s+', ' ', x).strip())
print("\nMenghapus spasi ganda")
data.head()
```

```python
data['content'] = data['content'].apply(lambda x: re.sub(r'^tempo\.co.*?- ', '', x, flags=re.IGNORECASE))
data['content'] = data['content'].apply(lambda x: re.sub(r'baca berita selengkapnya di sini\.', '', x, flags=re.IGNORECASE))
data['content'] = data['content'].apply(lambda x: re.sub(r'\bselanjutnya\b$', '', x, flags=re.IGNORECASE))
data['summary'] = data['summary'].apply(lambda x: re.sub(r'^tempo\.co.*?- ', '', x, flags=re.IGNORECASE))
data['summary'] = data['summary'].apply(lambda x: re.sub(r'baca berita selengkapnya di sini\.', '', x, flags=re.IGNORECASE))
data['summary'] = data['summary'].apply(lambda x: re.sub(r'\bselanjutnya\b$', '', x, flags=re.IGNORECASE))
print("\nMenghapus teks tidak relevan ('TEMPO.CO...', 'Baca berita selengkapnya di sini', atau 'Selanjutnya')")
data.head()
```

```python
# Ubah teks menjadi huruf kecil pada kolom 'content' dan 'summary'
data['content'] = data['content'].str.lower()
data['summary'] = data['summary'].str.lower()
print("\nMengubah teks menjadi huruf kecil")
data.head()
```

```python
# Tampilkan jumlah total data
total_data = len(data)
total_data
```

```python
import os

# Path untuk menyimpan file CSV
dataset_path = '/content/drive/MyDrive/!DEEP LEARNING/DATASET/clean_data.csv'

# Simpan dataset ke dalam file CSV
data.to_csv(dataset_path, index=False)

print(f"Dataset berhasil disimpan ke: {dataset_path}")
data.head()
```

```python
from sklearn.model_selection import train_test_split
train_data, temp_data = train_test_split(data, test_size=0.2, random_state=42)
val_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)
```

```python
print(f"Number of training samples: {len(train_data)}")
print(f"Number of validation samples: {len(val_data)}")
print(f"Number of test samples: {len(test_data)}")
```

```python
# Save datasets to CSV for download
dataset_path = '/content/drive/MyDrive/!DEEP LEARNING/DATASET'
train_data.to_csv(os.path.join(dataset_path, "train_data.csv"), index=False)
val_data.to_csv(os.path.join(dataset_path, "val_data.csv"), index=False)
test_data.to_csv(os.path.join(dataset_path, "test_data.csv"), index=False)
```
